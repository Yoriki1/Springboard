# Story from a dataset

Starbucks is the largest multinational coffee chain in the world. People love a good cup of coffee and drink from Starbucks, but do they know the details about their drink that they are ordering.

We will be investigating 'Nutrition facts for Starbucks Menu' from the Kaggle dataset. We will be asking questions to find out detail about the coffee chain's drinks. The questions that can be asked are:

* Which drinks has the most calories
* Which drinks has the most sugar
* Is there correlation between calories and sugar
* Which drinks has the most caffeine


## Importing Packages

We import the necessary packages to look at the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the dataset

We import the data and look at the contents of the data.

```python
df = pd.read_csv('starbucks_drinkMenu_expanded.csv')
```

```python
df.head()
```

```python
df.info()
```

## Looking into the data

We will create visualizations to better understand Starbuck's drinks

### Caloric distribution of Starbuck's drink

```python
calories = df['Calories']
_ = sns.histplot(calories)
plt.title('Calories Count')
plt.plot()
```

It looks like that most of the drinks at Starbuck's is around 200 calories

### Sugar (g) distribution of Starbuck's drinks

```python
sugar = df[' Sugars (g)']
_ = sns.histplot(sugar)
plt.title('Calories Count')
plt.plot()
```

It looks like most of Starbuck's Drinks has 30-45 grams of sugar.

### Calories by Beverage category

We will now ask the following questions: 
    
* Which drinks has the most calories
* Which drinks has the most sugar

```python
bar_cal = sns.barplot(x="Calories", y="Beverage_category", data=df)
plt.title('Calories by Beverage Category')
plt.plot()
```

We can see that smoothies and frappuccino blended coffee has the most amount of calories. Regular coffee has the least amount of calories.

### Sugar (g) by Beverage category

```python
bar_cal = sns.barplot(x=' Sugars (g)', y='Beverage_category', data=df)
plt.title('Sugar (g) by Beverage Category')
plt.plot()
```

We can see that frappucino blended coffee and frappuccino blended creme has the most amount of sugar. Regular coffee has the least amount of sugar.

### Relationship between Calories and Sugar (g)

Looking at the graphs above, there seems like there is a relationship between sugar (g) and calories. We will investigate this by graphing a scatter plot with sugar on the y-axis and calories on the x-axis.

```python
plt.figure(figsize=(15,15))
_ = sns.scatterplot(x = 'Calories', y = ' Sugars (g)', data = df, hue = 'Beverage_category', legend = 'auto',)
plt.title('Relationship Between Calories and Sugars')
plt.plot()
```

There seems to be a relationship between sugar and calories for Starbuck's drinks

### Calories by Beverage preperation

We will now look at the relationship between the beverage prep calories/sugar

```python
_ = sns.barplot(x = 'Calories', y='Beverage_prep', data=df)
plt.title('Calories by Beverage Preperation')
plt.plot()
```

### Sugar (g) by Beverage preperation

```python
_ = sns.barplot(x = ' Sugars (g)', y='Beverage_prep', data=df)
plt.title('Sugar (g) by Beverage Preperation')
plt.plot()
```

As expect whole milk drinks seems to have the most amount of calories and sugar

### Caffeine (mg) for Beverage category

#### Data Cleaning for Caffeine

We will preform some data cleaning for the caffeine category.

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

```python
df['Caffeine (mg)'] = df['Caffeine (mg)'].replace('Varies' , 0)
df['Caffeine (mg)'] = df['Caffeine (mg)'].replace('varies' , 0)
```

```python
df['Caffeine (mg)'] = pd.to_numeric(df['Caffeine (mg)'])
```

###  Graph for Caffeine (mg) for Beverage category

```python
_ = sns.barplot(x='Caffeine (mg)', y="Beverage_category", data=df)
plt.title('Caffeine (mg) for Beverage Category')
plt.plot()
```

It is no suprise that coffee has the most amount of caffeine out of all of the drink categories
